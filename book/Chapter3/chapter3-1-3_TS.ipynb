## Derivation of strong form 

**Already material here (currently in chapter3-1_1D_Problems.ipynb)**

Specifically : page 82, Equations 6.35 - 6.40

The Timoshenko beam allows shear deformation and is more suitable for relatively short beams. In Timoshenko beam theory, planes remain plain during deformation but are not needed to remain normal to the longitudinal axis. This assumption allows for shear deformation. The plane is assumed to have a rotation $\theta$ of aplane given by the equation

$$
\theta = \frac{dv}{dx} - \gamma
$$ (TS_KIN_eq1)

where $\gamma$ is the rotation due to shear. In terms of the shear force and properties of the beam, it is related to the shear force by the following constitutive relationship:

$$
\gamma = \frac{q}{GA_s}
$$ (TS_SHEAR_eq1)

where $G$ is the shear modulus and $A_s$ is the effective shear area. Relative to the Euler-Bernoulli theory, an additional unknown, the shear strain $\gamma$, has been introduced. The governing equations are given by the equilibrium conditions in equations {numref}`beam_rot_eq3` and {numref}`beam_tran_eq3` with the following constitutive relationships:

$$
m = -EI\kappa
$$ (TS_CONST_eq1)


$$
q = GA_s \gamma = GA_s (\frac{dv}{dx} - \theta)
$$ (TS_CONST_eq2)

and considering $v$ and $\theta$ to be the fundamental unknowns. Inserting the constitutive relationships into the equilibrium equations {numref}`beam_rot_eq3` and {numref}`beam_tran_eq3` leads to:

$$
-EI \frac{d^2 \theta}{dx^2} - GA_s (\frac{dv}{dx} - \theta) = 0
$$ (TS_STRONG_eq1)


$$
GA_s (\frac{d^2 v}{dx^2} - \frac{d\theta}{dx}) + f_y = 0
$$ (TS_STRONG_eq2)

which are two coupled second-order equations, the first for rotation $\theta$, and the second for the displacement $v$, in the space $\Omega$. Together with the previously defined boundary conditions, this problem is complete.

## Derivation of weak form 
**Weak governing equations**

Multiplying equations {numref}`TS_STRONG_eq1` and {numref}`TS_STRONG_eq2` by appropriately defined weight functions $\overline{\theta}$ and $\overline{v}$,

$$
- \int_{\Omega} \overline{\theta} EI \theta_{,xx} \: d\Omega - \int_{\Omega} \overline{\theta} GA_s (v_{,x} - \theta) \: d\Omega = 0
$$ (TS_WF_eq1)

$$
\int_{\Omega} \overline{v} GA_s (v_{,xx} - \theta_{,x}) \: d\Omega + \int_{\Omega} \overline{v} f_y \: d\Omega = 0
$$ (TS_WF_eq2)

Applying integrations by parts once gives

$$
\int_{\Omega} \overline{\theta}_{,x} EI \theta_{,x} \: d\Omega - \int_{\Omega} \overline{\theta} GA_s (v_{,x} - \theta) \: d\Omega + \int_{\Gamma_M} \overline{\theta} T \: d\Gamma = 0
$$ (TS_WF_eq3)

$$
- \int_{\Omega} \overline{v}_{,x} GA_s (v_{,x} - \theta) \: d\Omega + \int_{\Gamma} \overline{v} GA_s (v_{,x} - \theta) n_x \: d\Gamma  + \int_{\Omega} \overline{v} f_y \: d\Omega = 0
$$ (TS_WF_eq4)

and then inserting the Neumann boundary conditions yields the weak problem for a Timoshenko beam of: find $v \in S_v$ and $\theta \in S_{\theta}$ such that

$$
\int_{\Omega} \overline{\theta}_{,x} EI \theta_{,x} \: d\Omega - \int_{\Omega} \overline{\theta} GA_s (v_{,x} - \theta) \: d\Omega + \int_{\Gamma_M} \overline{\theta} T \: d\Gamma = 0
$$ (TS_WF_eq5)


$$
- \int_{\Omega} \overline{v}_{,x} GA_s (v_{,x} - \theta) \: d\Omega + \int_{\Gamma_Q} \overline{v} F_y \: d\Gamma + \int_{\Omega} \overline{v} f_y \: d\Omega = 0
$$ (TS_WF_eq6)

A more precise definition of the necessary functions spaces should be provided to complete the problem.

The finite element formulation for the Timoshenko beam is relatively simple, due tp the second order nature of the governing equations. This will allow the use of the standard shape functions. 

The Galerkin problem for the Timoshenko beam involves finding : 
$$ v^h ∈ S_v^h $$
$$  \theta ^h  ∈ S_θ^h  $$

where $$ S_h ⊂ S $$ and $$ V_h ⊂ V $$ are finite dimensional spaces, 

such that 

$$ \int_{Ω} \overline{θ}_{,x}^h EI θ_{,x}^h  dΩ
-\int_{Ω} \overline{θ}_{,x}^h GA_s θ_{,x}^h  dΩ = \int_{Ω} \overline{θ}^hT  dΓ
$$(6.62)


$$ \int_{Ω} \overline{u}_{,x}^h GA_s ( u_{,x}^h - θ^h) dΩ =  
\int_{Ω} \overline{u}^h f_y dΩ + \int_{Γ_{Q}} \overline{u}^h F_y  dΓ
$$(6.63)

Considering the representation of the fields $u^h$ and $θ^h$ and corresponding derivatives i nterms of shape functions and nodal avariable, 

$$
u^h = \textbf{N} ^u \textbf{α}_e^u
$$ (6.64a)

$$
\theta ^h  = \textbf{N} ^θ  \textbf{α}_e^θ 
$$ (6.64b)
$$
\overline{v}^h = \textbf{N} ^u \textbf{b}_e^u 
$$ (6.64c)
$$
\overline{\theta} ^h  = \textbf{N} ^θ  \textbf{b}_e^θ 
$$ (6.64d)

A distinction is made between the shape functions for $u^h$ and $\theta^h$

as these may differ. Inserting the expressions for the unknown fields in terms of nodal variables and variations into equations {numref}`6.62` and {numref}`6.63`  leads to: 

$$  
\int_{Ω_e}\textbf{Β}^{θ^Τ} ΕΙ \textbf{Β}^θ  dΩ \textbf{α}_e^θ+
\int_{Ω_e}\textbf{N}^{θ^Τ} GA_s \textbf{N}^θ  dΩ \textbf{α}_e^θ +
\int_{Ω_e}\textbf{N}^{θ^Τ} GA_s \textbf{Β}^u  dΩ \textbf{α}_e^u = - 
\int_{Γ_{e,M}} \textbf{N}^{θ^Τ}  Τ dΓ 
$$(6.65)


$$
\int_{Ω_e}\textbf{Β}^{u^Τ} GA_s \textbf{Β}^u  dΩ \textbf{α}_e^u-
\int_{Ω_e}\textbf{B}^{u^Τ} GA_s \textbf{N}^θ  dΩ \textbf{α}_e^θ =
\int_{Γ_{e,Q}}\textbf{N}^{u^Τ} F_y  dΓ +
\int_{Ω_e} \textbf{N}^{u^Τ}  f_y dΩ
$$(6.66)

The stiffness matrix of an element is of the form 

$$ \begin{bmatrix}   f_θ \\ f_u \end{bmatrix}  = \begin{bmatrix}   \textbf{k}^{θθ} & \textbf{k}^{θu} \\
  \textbf{k}^{uθ} & \textbf{k}^{uu}  \\ \end{bmatrix} *   \begin{bmatrix}   α_θ \\ α_u \end{bmatrix}
$$(6.67)

where the components of the the element stiffness matrix are given by
$$ 
\textbf{k}^{θθ} =  \int_{Ωe} \textbf{B}^{θ^Τ} EI \textbf{B}^{θ} \textbf{N}^{θ^Τ} G A_s \textbf{N}^{θ} dΩ 
$$ (6.68)
$$
 \textbf{k}^{θu} =  \int_{Ωe} \textbf{N}^{θ^Τ} G A_s \textbf{B}^{u} dΩ  
 $$ (6.69)
$$
 \textbf{k}^{uθ} =  \int_{Ωe} \textbf{B}^{u^Τ} G A_s \textbf{N}^{θ} dΩ  
 $$ (6.70)
$$
 \textbf{k}^{uu} =  \int_{Ωe} \textbf{B}^{u^Τ} G A_s \textbf{B}^{u} dΩ  
$$ (6.71)

and components of the element RHS vector are given by 

$$
 \textbf{f} ^{θ}  = \int_{Γ_{e,Μ}} \textbf{N}^{θ^Τ} T dΓ 
$$ (6.72)
$$
  \textbf{f} ^{u} = \int_{Γ_{e,Q}} \textbf{N}^{u^Τ}  F_y dΓ + \int_{Ω_e}\textbf{N}^{u^Τ} f_y  dΩ  
$$ (6.73)

## Locking Phenomena 

Add content from pages 88-90

This chapter presents the topic of shear locking. As seen in previous chapters, Timoshenko beam theory is valid for both short and slender beams. Conceptually, only a Timoshenko beam element would suffice for the analysis of both short and slender elements. This would be advantageous from a point of view of generality and simplicity as the Timoshenko beam uses $ C^0 $ shape functions.  However, when applying Timoshenko elements for thin, bending problems the results are influenced by the so-called **shear locking**. 


<div class="alert alert-block alert-info">
<b>Note:</b>  
Shear locking is the effect that occues as a result of the finite element analysis, in which shear contribution to the energy does not vanish, resulting in an overly stiff response and bending displacements smaller than they should be. This practically means that the element cannot capture the kinematics of the deformation.
</div>


Now, let's consider a Timoshenko beam of length L, with linear shape functions for both displacement and rotation. The components of the stiffness matrix are given by:


$$ k_e^{θθ} = \begin{bmatrix} \frac{EI}{L} + \frac{GA_sL}{3} & \frac{-EI}{L} + \frac{GA_sL}{6}  \\ \frac{-EI}{L} + \frac{GA_sL}{6} & \frac{EI}{L} + \frac{GA_sL}{3}   \end{bmatrix}$$

$$ k_e^{θu} = \begin{bmatrix} \frac{-EI}{L} +\frac{GA_s}{6} & \frac{EI}{L} +\frac{GA_s}{3}  \\  \frac{GA_s}{2} &  \frac{-GA_s}{2}   \end{bmatrix}$$

$$ k_e^{uu} = \begin{bmatrix} \frac{GA_s}{L} &  \frac{-GA_s}{L}  \\  \frac{-GA_s}{L} &  \frac{GA_s}{L}   \end{bmatrix}$$

Therefore, the stiffness matrix is of the form

$$ k_e = \begin{bmatrix} \frac{EI}{L} + \frac{GA_sL}{3} &  \frac{-EI}{L} + \frac{GA_sL}{6} & \frac{GA_s}{2} &  \frac{-GA_s}{2}  \\
 \frac{-EI}{L} + \frac{GA_sL}{6} &   \frac{EI}{L} + \frac{GA_sL}{3}  & \frac{GA_s}{2} &  \frac{-GA_s}{2} \\
  \frac{GA_s}{2} &  \frac{GA_s}{2}  & \frac{GA_s}{L} &  \frac{-GA_s}{L}  \\
  \frac{-GA_s}{2} &  \frac{-GA_s}{2}  &  \frac{-GA_s}{L} &  \frac{GA_s}{L}\\
  \end{bmatrix}$$



If the one end of the beam is fixed, so that u1=θ1=0 and a shear load P is applied at the other end, the problem is then the following


$$ \begin{bmatrix}   0 \\ P \end{bmatrix}  = \begin{bmatrix} \frac{EI}{L} + \frac{GA_sL}{3} &  \frac{-GA_s}{2}  \\
  \frac{-GA_s}{2} &  \frac{GA_s}{L}    \\ \end{bmatrix} *  \begin{bmatrix}   θ_2 \\ u_2 \end{bmatrix} 

$$

Finally, the solution for the displacement $u_2$ is 

$$ u_2 = \frac{P( \frac{EI}{L} + \frac{GA_sL}{3} ) } {\frac{EIGA_s}{L^2}+\frac{G^2A_s^2}{12}}   $$



<div class="alert alert-block alert-info">
<b>Note:</b> 

Two limit states can now be identified: 

For L -> $\infty$ : $$ u_2 \approx \frac {4 PL}{G A_s} $$

which is independent of the moment of inertia I and will yield a very stiff response.

For L -> 0 : $$ u_2 \approx \frac {PL}{GA_s} $$ 

with the latter being the correct result. Even with an extremely fine mesh, thie element will exhibit a very stiff response.
</div>

For the case of **one-point numerical integration** of the term $$ \int_{Ωe} N^{θ^Τ}GA_sN^{θ^Τ} dΩ $$

the element stiffness matrix would be:

$$ k_e = \begin{bmatrix} \frac{EI}{L} + \frac{GA_sL}{4} &  \frac{-EI}{L} + \frac{GA_sL}{4} & \frac{GA_s}{2} &  \frac{-GA_s}{2}  \\
 \frac{-EI}{L} + \frac{GA_s}{4} &   \frac{EI}{L} + \frac{GA_sL}{4}  & \frac{GA_s}{2} &  \frac{-GA_s}{2} \\
  \frac{GA_s}{2} &  \frac{GA_s}{2}  & \frac{GA_s}{L} &  \frac{-GA_s}{L}  \\
  \frac{-GA_s}{2} &  \frac{-GA_s}{2}  &  \frac{-GA_s}{L} &  \frac{GA_s}{L}\\
  \end{bmatrix}$$
