# Timoshenko Beam

## Derivation of strong form 

The Timoshenko beam allows shear deformation and is valid for slender as well as for relatively short beams. In Timoshenko beam theory, planes normal to the beam axis remain plain but do not necessarily remain normal to the longitudinal axis. The rotation of the normal to the beam axis $\theta$ is not directly coupled to the first derivative of the deflection $\frac{dw}{dx}$. This assumption allows for shear deformation, where the shear strain $\gamma$ is a measure for how much the 

$$
\gamma = \frac{dw}{dx} - \theta
$$ (TS-KIN-eq1)

Is it apparent from {numref}`TS-KIN-eq1` that assuming $\gamma$ to be equal to 0 implies a direct relation between $w$ and $\theta$ after which and only one unknown field remains: when $w(x)$ is known, $\theta$ follows, as is the case in the Euler-Bernouilli beam theory. In Timoshenko beam theory, however this is not the case. Shear deformation can be unequal to zero, and as a consequence $w$ and $\theta$ are independent fields. The two fields will be coupled through the shear stiffness, but in the finite element formulation they will give rise to separate degrees of freedom. 

$$
\gamma = \frac{q}{GA_s}
$$ (TS_SHEAR_eq1)

where $G$ is the shear modulus and $A_s$ is the effective shear area. Relative to the Euler-Bernoulli theory, an additional unknown, the shear strain $\gamma$, has been introduced. The governing equations are given by the equilibrium conditions in equations {numref}`beam_rot_eq3` and {numref}`beam_tran_eq3` with the following constitutive relationships:

$$
m = -EI\kappa
$$ (TS_CONST_eq1)


$$
q = GA_s \gamma = GA_s \left(\frac{dw}{dx} - \theta\right)
$$ (TS_CONST_eq2)

and considering $w$ and $\theta$ to be the fundamental unknowns. Inserting the constitutive relationships into the equilibrium equations {numref}`beam_rot_eq3` and {numref}`beam_tran_eq3` leads to:

$$
-EI \frac{d^2 \theta}{dx^2} - GA_s \left(\frac{dw}{dx} - \theta\right) = 0
$$ (TS_STRONG_eq1)


$$
GA_s \left(\frac{d^2 w}{dx^2} - \frac{d\theta}{dx}\right) + f_y = 0
$$ (TS_STRONG_eq2)

which are two coupled second-order equations, the first for rotation $\theta$, and the second for the displacement $w$, in the space $\Omega$. Together with the previously defined boundary conditions, this problem is complete.

## Derivation of weak form 
**Weak governing equations**

Multiplying equations {numref}`TS_STRONG_eq1` and {numref}`TS_STRONG_eq2` by appropriately defined weight functions $\overline{\theta}$ and $\overline{w}$,

$$
- \int_{\Omega} \overline{\theta} EI \theta_{,xx} \: d\Omega - \int_{\Omega} \overline{\theta} GA_s (w_{,x} - \theta) \: d\Omega = 0
$$ (TS_WF_eq1)

$$
\int_{\Omega} \overline{w} GA_s (w_{,xx} - \theta_{,x}) \: d\Omega + \int_{\Omega} \overline{w} f_y \: d\Omega = 0
$$ (TS_WF_eq2)

Applying integrations by parts once gives

$$
\int_{\Omega} \overline{\theta}_{,x} EI \theta_{,x} \: d\Omega - \int_{\Omega} \overline{\theta} GA_s (w_{,x} - \theta) \: d\Omega + \int_{\Gamma_M} \overline{\theta} T \: d\Gamma = 0
$$ (TS_WF_eq3)

$$
- \int_{\Omega} \overline{w}_{,x} GA_s (w_{,x} - \theta) \: d\Omega + \int_{\Gamma} \overline{w} GA_s (w_{,x} - \theta) n_x \: d\Gamma  + \int_{\Omega} \overline{w} f_y \: d\Omega = 0
$$ (TS_WF_eq4)

and then inserting the Neumann boundary conditions yields the weak problem for a Timoshenko beam of: find $w \in S_w$ and $\theta \in S_{\theta}$ such that

$$
\int_{\Omega} \overline{\theta}_{,x} EI \theta_{,x} \: d\Omega - \int_{\Omega} \overline{\theta} GA_s (w_{,x} - \theta) \: d\Omega + \int_{\Gamma_M} \overline{\theta} T \: d\Gamma = 0
$$ (TS_WF_eq5)


$$
- \int_{\Omega} \overline{w}_{,x} GA_s (w_{,x} - \theta) \: d\Omega + \int_{\Gamma_Q} \overline{w} F_y \: d\Gamma + \int_{\Omega} \overline{w} f_y \: d\Omega = 0
$$ (TS_WF_eq6)

A more precise definition of the necessary functions spaces should be provided to complete the problem. 



## Matrix Form 

The finite element formulation for the Timoshenko beam is relatively simple, due to the second order nature of the governing equations. This will allow the use of the standard shape functions. 

The Galerkin problem for the Timoshenko beam involves finding : 
$$ w^h ∈ S_w^h $$
$$  \theta ^h  ∈ S_\theta^h  $$

such that 

$$ \int_{\Omega} \overline{\theta}_{,x}^h EI \theta_{,x}^h  d\Omega
-\int_{\Omega} \overline{\theta}_{,x}^h GA_s (\overline{w}_{,x}^h- \theta^h)  d\Omega = 
\int_{\Omega} \overline{\theta}^hT  d\Gamma
$$(6.62)


$$ \int_{\Omega} \overline{w}_{,x}^h GA_s ( w_{,x}^h - \theta^h) d\Omega =  
\int_{\Omega} \overline{w}^h f_y d\Omega + \int_{\Gamma_{Q}} \overline{w}^h F_y  d\Gamma
$$(6.63)

Considering the representation of the fields $w^h$ and $\theta^h$ and corresponding derivatives i nterms of shape functions and nodal avariable, 

$$
w^h = \textbf{N} ^w \textbf{a}_e^w
$$ (6.64a)

$$
\theta ^h  = \textbf{N} ^\theta  \textbf{a}_e^\theta 
$$ (6.64b)
$$
\overline{w}^h = \textbf{N} ^w \textbf{b}_e^w 
$$ (6.64c)
$$
\overline{\theta} ^h  = \textbf{N} ^\theta  \textbf{b}_e^\theta 
$$ (6.64d)

A distinction is made between the shape functions for $w^h$ and $\theta^h$

as these may differ. Inserting the expressions for the unknown fields in terms of nodal variables and variations into equations {numref}`6.62` and {numref}`6.63`  leads to:

$$
\int_{\Omega_e}\textbf{B}_{w}^T GA_s \textbf{B}_w  d\Omega \textbf{a}_e^w-
\int_{\Omega_e}\textbf{B}_{w}^T GA_s \textbf{N}_\theta  d\Omega \textbf{a}_e^\theta =
\int_{\Gamma_{e,Q}}\textbf{N}_{w}^T F_y  d\Gamma +
\int_{\Omega_e} \textbf{N}_{w}^T  f_y d\Omega
$$(6.66)

The stiffness matrix of an element is of the form 

$$ \begin{bmatrix}   \textbf{f}_\theta \\ \textbf{f}_w \end{bmatrix}  = \begin{bmatrix}   \textbf{K}_{\theta\theta} & \textbf{K}_{\theta w} \\
  \textbf{K}_{w\theta} & \textbf{K}_{ww}  \\ \end{bmatrix} *   \begin{bmatrix}   \textbf{a}_\theta \\ \textbf{a}_w \end{bmatrix}
$$(6.67)

where the components of the the element stiffness matrix are given by
$$

$$ (6.68)
$$
 \textbf{K}_{\theta w} =  -\int_{\Omega_e} \textbf{N}_{\theta}^T G A_s \textbf{B}_{w} d\Omega  
 $$ (6.69)
$$
 \textbf{K}_{w\theta} =  -\int_{\Omega_e} \textbf{B}_{w}^T G A_s \textbf{N}_{\theta} d\Omega  
 $$ (6.70)
$$

$$ (6.71)

$$

## Locking Phenomena

This chapter presents the topic of shear locking. As seen in previous chapters, Timoshenko beam theory is valid for both short and slender beams. Conceptually, only a Timoshenko beam element would suffice for the analysis of both short and slender elements. This would be advantageous from a point of view of generality and simplicity as the Timoshenko beam uses $ C^0 $ shape functions.  However, when applying Timoshenko elements for thin, bending problems the results are influenced by the so-called **shear locking**. 


<div class="alert alert-block alert-info">
<b>Note:</b>  
Shear locking is the effect that occues as a result of the finite element analysis, in which shear contribution to the energy does not vanish, resulting in an overly stiff response and bending displacements smaller than they should be. This practically means that the element cannot capture the kinematics of the deformation.
</div>


Now, let's consider a Timoshenko beam of length L, with linear shape functions for both displacement and rotation. The components of the stiffness matrix are given by:


$$ \mathbf{K}^e_{\theta\theta} = \begin{bmatrix} \frac{EI}{L} + \frac{GA_sL}{3} & \frac{-EI}{L} + \frac{GA_sL}{6}  \\ \frac{-EI}{L} + \frac{GA_sL}{6} & \frac{EI}{L} + \frac{GA_sL}{3}   \end{bmatrix}$$

$$ \mathbf{K}^e_{\theta w} = \begin{bmatrix} \frac{-EI}{L} +\frac{GA_s}{6} & \frac{EI}{L} +\frac{GA_s}{3}  \\  \frac{GA_s}{2} &  \frac{-GA_s}{2}   \end{bmatrix}$$

$$ \mathbf{K}^e_{ww} = \begin{bmatrix} \frac{GA_s}{L} &  \frac{-GA_s}{L}  \\  \frac{-GA_s}{L} &  \frac{GA_s}{L}   \end{bmatrix}$$

Therefore, the stiffness matrix is of the form

$$ \mathbf{K}^e = \begin{bmatrix} \frac{EI}{L} + \frac{GA_sL}{3} &  \frac{-EI}{L} + \frac{GA_sL}{6} & \frac{GA_s}{2} &  \frac{-GA_s}{2}  \\
 \frac{-EI}{L} + \frac{GA_sL}{6} &   \frac{EI}{L} + \frac{GA_sL}{3}  & \frac{GA_s}{2} &  \frac{-GA_s}{2} \\
  \frac{GA_s}{2} &  \frac{GA_s}{2}  & \frac{GA_s}{L} &  \frac{-GA_s}{L}  \\
  \frac{-GA_s}{2} &  \frac{-GA_s}{2}  &  \frac{-GA_s}{L} &  \frac{GA_s}{L}\\
  \end{bmatrix}$$



If the one end of the beam is fixed, so that $w_1=\theta_1=0$ and a shear load $F_y$ is applied at the other end, the problem is then the following


$$ \begin{bmatrix} \frac{EI}{L} + \frac{GA_sL}{3} &  \frac{-GA_s}{2}  \\
  \frac{-GA_s}{2} &  \frac{GA_s}{L}   \end{bmatrix}  \begin{bmatrix}   \theta_2 \\ w_2 \end{bmatrix} = \begin{bmatrix}   0 \\ F_y \end{bmatrix} 

$$

Finally, the solution for the displacement $w_2$ is 

$$ w_2 = \frac{F_y( \frac{EI}{L} + \frac{GA_sL}{3} ) } {\frac{EIGA_s}{L^2}+\frac{G^2A_s^2}{12}}   $$



<div class="alert alert-block alert-info">
<b>Note:</b> 

Two limit states can now be identified: 

For L -> $\infty$ : $$ w_2 \approx \frac {4 PL}{G A_s} $$

which is independent of the moment of inertia I and will yield a very stiff response.

For L -> 0 : $$ w_2 \approx \frac {PL}{GA_s} $$ 

with the latter being the correct result. Even with an extremely fine mesh, thie element will exhibit a very stiff response.
</div>

$$ \mathbf{K}^e = \begin{bmatrix} \frac{EI}{L} + \frac{GA_sL}{4} &  \frac{-EI}{L} + \frac{GA_sL}{4} & \frac{GA_s}{2} &  \frac{-GA_s}{2}  \\
 \frac{-EI}{L} + \frac{GA_s}{4} &   \frac{EI}{L} + \frac{GA_sL}{4}  & \frac{GA_s}{2} &  \frac{-GA_s}{2} \\

  \end{bmatrix}$$
